### There are 891 Rows and 7 Columns in the dataset.

```python
data = data.drop(columns= ['Unnamed: 0'])
```

```python
data
```

### We have dropped the unnamed column from the dataset

```python
data.index = data.index + 1
```

```python
data
```

### We have set the index accordingly.

```python
data['course_students_enrolled'] = data['course_students_enrolled'].str.replace('k', '*1000')
data['course_students_enrolled'] = data['course_students_enrolled'].str.replace('m', '*1000000')
data['course_students_enrolled'] = data['course_students_enrolled'].map(lambda x: eval(x))
data['course_students_enrolled']
```

### We have converted the values present in the 'course_students_enrolled' as 5.3k to 5300.0

```python
data
```

## As we have completed the data pre-processing we can start analysing the data.

## Types of Courses available on Coursera

```python
plt.figure(figsize=(30,20))
sns.countplot(data['course_Certificate_type'])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('type of certificate', fontsize=30)
plt.ylabel('count', fontsize=30)
plt.title("TYPE OF COURSE", fontsize=40)

plt.show()
```

## From the above graph we can conclude that 'course' type is maximum followed by specialization and professional certificate.

## COURSE DIFFICULTY LEVEL OF COURSES

```python
data
```

```python
df= data.course_difficulty.value_counts()
```

```python
df
```

```python
plt.figure(figsize=(10,20))
data['course_difficulty'].value_counts().plot.pie(autopct= '%1.1f%%')
plt.title("COURSE DIFFICULTY",fontsize=3)
plt.show()
          
          
          
```

## From the above pie chart it is clear that most of the courses are beginner level. Intermediate and mixed levels are almost equal in share and Advanced level is of very less percentage.


## Top 10 Organizations based on courses provided

```python
data
```

```python
df1 = data.groupby(['course_organization'])['course_organization'].count().to_dict()
df1 = pd.DataFrame({'course_organization': list(df1.keys()), 'Count': list(df1.values())})
df1 = df1.sort_values(by=['Count'], ascending=False)
df1 = df1.head(10)
df1
```

```python
plt.figure(figsize=(30,20))
plt.bar(df1['course_organization'], df1['Count'] ,color='dimgrey')

plt.xticks(rotation = 90,fontsize= 30)
plt.yticks(fontsize= 20)
plt.xlabel('organization' ,fontsize=40)
plt.ylabel('count',fontsize=40)
plt.title("TOP 10 ORGANIZATIONS (COURSES-WISE)",fontsize=40)

plt.show()
```

## We have a clear view of the above graph which shows that the University of Pennsylvania has contributed with the maximum number of courses.


## Top 10 Organizations based on number of students enrolled.

```python
df2 = data.loc[:, ['course_organization','course_students_enrolled']]
df2 = pd.DataFrame(df2.groupby(['course_organization'])['course_students_enrolled'].sum())
df2 = df2.sort_values(by=['course_students_enrolled'], ascending=False)
df2 = df2.head(10)
df2
df2.reset_index()
```

```python
plt.figure(figsize=(10,20))
df2['course_students_enrolled'].plot.pie(autopct= '%1.1f%%' , ylabel = '')
plt.title("TOP 10 ORGANIZATIONS(STUDENTS ENROLLED)" , fontsize = 20)

plt.show()
```

## We can see from the above pie chart that University of Michigan has got the maximum number of students enrolled in their courses followed by University of Pennsylvania, Stanford University and others .

## Number of students enrolled Vs Course Rating

```python
data.groupby('course_rating')['course_students_enrolled'].count().plot(kind='barh', grid=True,
    figsize=(15, 10)).set_xlabel('Students enrolled',fontsize=20)
plt.title("STUDENTS ENROLLED Vs COURSE RATING", fontsize = 30)
plt.show()
```

## As we have plotted a horizontal bar graph we can see that maximum number of students have given a rating of 4.8 follwed by 4.7 for the courses on the whole.

## Organization Vs Maximum Course Rating 

```python
df3 = round(data.groupby(['course_organization'])['course_rating'].mean()).to_dict()
df3 = pd.DataFrame({'course_organization': list(df3.keys()), 'course_rating': list(df3.values())})
df_1 = df3[df3['course_rating'] >=4.9]
df_1
```

```python
df_2= df_1.head(10)
df_2
```

## The above table shows a few universities/organizations which have recieved the maximum course rating i.e,5.0


## Course Rating among all the courses

 

```python
plt.figure(figsize=(20,20))
sns.displot(data=data.course_rating,height = 10,kde=True)
plt.title("COURSE RATING AMONG ALL COURSES",fontsize=30)
plt.show()
```

## On the whole lot of courses available, maximum have received a rating of 4.8

## Top 10 Courses based on number of students enrolled.

```python
top10_courses = data.loc[:,['course_title', 'course_students_enrolled']]
top10_courses = top10_courses.sort_values(by=['course_students_enrolled'], ascending=False)
top10_courses = top10_courses.head(10)
top10_courses
```

```python
fig = plt.figure(figsize=(16, 12))
ax = fig.add_axes([0,0,1,1])
ax.bar(top10_courses['course_title'], top10_courses['course_students_enrolled'], color="#ff6361")
plt.xticks(rotation = 90,fontsize= 20)
plt.yticks(fontsize= 20)
plt.xlabel('COURSE NAME' ,fontsize=20)
plt.ylabel('count of students',fontsize=20)
plt.title("TOP 10 COURSES (STUDENTS ENROLLED)",fontsize=40)

plt.show()
```

## We can see that Machine Learning is the top course in coursera with maximum students enrolled.
